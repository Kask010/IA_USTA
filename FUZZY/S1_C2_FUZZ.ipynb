# Importando librerias

```python
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import numpy as np
import matplotlib.pyplot as plt
```

# Ejercicio 1: Graficar variable temperatura

```python
Temp=ctrl.Antecedent(np.arange(80,110,1),'Temperatura')
Temp['High']=fuzz.trapmf(Temp.universe,[80,80,90,100])
Temp['Very High']=fuzz.trapmf(Temp.universe,[90,100,110,110])
Temp.view()
```

# Ejercicio 2: Tres funciones de pertenencia

```python
Temp=ctrl.Antecedent(np.arange(60,110,1),'Temperatura')
Temp['Mid']=fuzz.trapmf(Temp.universe,[60,60,70,80])
Temp['High']=fuzz.trapmf(Temp.universe,[70,80,90,100])
Temp['Very High']=fuzz.trapmf(Temp.universe,[90,100,110,110])
Temp.view()
```
